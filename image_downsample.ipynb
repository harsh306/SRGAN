```python
## Aim of this exersice is to show what does it mean to downsample or low-resolution images from high resolution images. And what happens when we try to upsace them directly
```

```python
from PIL import Image
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = "/Users/hpathak/Downloads/Golden-Snitch-Harry-Potter-Vinyl-Decal-Sticker.jpg"
```

```python
#orignal image
img = Image.open(img_path)
img_a =np.array(img)
print(img_a.shape)
plt.imshow(img_a)
```

```python
#Low resoltuion 
size = 50,50
img_resized = img.resize(size, Image.ANTIALIAS)
img_resiezed_a =np.array(img_resized)
print(img_resiezed_a.shape)
plt.imshow(img_resiezed_a)
```

```python
#back to high resolution
size = 1000,1000
img_ = img_resized.resize(size, Image.ANTIALIAS)
img_a_ =np.array(img_)
print(img_a_.shape)
plt.imshow(img_a_)
```
